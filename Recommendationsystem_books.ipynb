### no.of unique books

```python
# Changing data set to matrix form using pivot table
book_df=book.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
book_df
```

```python
# Setting index to userid's
book_df.index=b
book_df
```

```python
# imputing null values to 0
book_df.fillna(0, inplace=True)
book_df
```

```python
# Calculating cosine similarity 
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
book_sim = 1 - pairwise_distances( book_df.values,metric='cosine')
```

```python
book_sim
```

```python
# Storing them in a dataframe
book_sim_df = pd.DataFrame(book_sim)
book_sim_df
```

```python
# Filling diagonal values to 0 and setting index ,column names to userid's
np.fill_diagonal(book_sim,0)
book_sim_df.columns=b
book_sim_df.index=b
book_sim_df
```
